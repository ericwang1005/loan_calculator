```python
totalAmount = 12000000
amount = totalAmount
years = 30
rate = 2.06
period = years*12
month_rate = rate/100/12
month_pay = int(amount/period)

datas = []
totalInterest = 0
for i in range(period):
    interest = round(amount*month_rate)
    amount -= month_pay
    # 最後一期?
    if i ==period-1:
        datas.append([i+1,month_pay+amount,interest,month_pay+amount+interest,0])
        print(f'期別:{i+1} 本金:{month_pay+amount} 利息:{interest} 本利和:{month_pay+amount+interest} 餘額:{0}')
    else:
        datas.append({i+1,month_pay,interest,month_pay+interest,amount})
        print(f'期別:{i+1} 本金:{month_pay} 利息:{interest} 本利和:{month_pay+interest} 餘額:{amount}')

    totalInterest+=interest
print(f'總還款金額:{totalAmount+totalInterest} 總利息支出:{totalInterest}')
```
